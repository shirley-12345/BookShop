# Preprocessing

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tqdm import tqdm


##############show the picture##############

#set the data directory
TrainValidateDATADIR = "/content/drive/MyDrive/Canada_percentage/Scalogram/"


CATEGORIES = ["50%", "100%", "150%", "190%"]

for category in CATEGORIES:
    path = os.path.join(TrainValidateDATADIR, category)
    for img in os.listdir(path):
      image_array = cv2.imread(os.path.join(path, img))
      plt.imshow(image_array,cmap = 'turbo')
      plt.show()
      break
    break

```

```python
import random
import numpy as np


##############build the training data##############
training_data = []
def create_training_data():
  for category in CATEGORIES:
    path = os.path.join(TrainValidateDATADIR, category)    # join the directory and category, such as /content/drive/MyDrive/Canada/scalogram/hydrated
    class_num = CATEGORIES.index(category)    # 0:hydrated; 1:dehydrated
    for img in tqdm(os.listdir(path)):        # iterate through all the images in the directory; tqdm: generate progress bar
      try:
        img_array = cv2.imread(os.path.join(path, img))     #read the data
        print(img_array)
        training_data.append([img_array, class_num])        #append the data to traning data
      except Exception as e:
        pass

create_training_data()
print('\n',len(training_data),'\n')



##############shuffle the traning data##############
random.shuffle(training_data)

trainX = []
trainy = []

for features,label in training_data:
    trainX.append(features)
    trainy.append(label)


print(len(trainX),len(trainy))
print(trainy)

##############define one hot##############
def indices_to_one_hot(data, nb_classes):
    targets = np.array(data).reshape(-1)
    return np.eye(nb_classes)[targets]

#print(indices_to_one_hot(trainy,4))
##############data spilt into train and validation##############
scalogram = np.array(trainX)
label = np.array(indices_to_one_hot(trainy,4))

train_data, valid_data= np.split(scalogram,[int(0.8*len(scalogram))])
train_label, valid_label= np.split(label,[int(0.8*len(label))])
print("training", train_label)
print(len(train_label))
print(valid_label)
```

# Small scale CNN model

```python
##############customize CNN##############
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

model = models.Sequential()
model.add(layers.Conv2D(8, kernel_size=(3, 3), activation='relu', input_shape=(480, 640, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(16, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(256, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(256, (3, 3), activation='relu'))


model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(4))

model.summary()

model.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(train_data, train_label, batch_size=8, epochs=10)

print(history)

plt.plot(history.history['accuracy'], label='Training accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')

test_loss, test_acc = model.evaluate(valid_data,  valid_label, batch_size=8, verbose=1)
```
